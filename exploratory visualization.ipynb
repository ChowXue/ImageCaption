```python
import sys
import pickle
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
import numpy as np
from tqdm import tqdm
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
file_name_caption = pickle.load(open( "file_name_caption.bat", "rb" ))
```

```python
lenghs = []
sents = []
words = Counter()
for caption in tqdm(file_name_caption.values(), file=sys.stdout, total=len(file_name_caption)):
    sent = word_tokenize(caption)
    sents.append(sent)
    lenghs.append(len(sent))
    words.update(list(sent))
```

```python
df = pd.DataFrame()
df['lengths'] = lenghs
```

```python
df.plot(kind='hist', bins=100)
```

```python
np.max(lenghs)
```

```python
np.min(lenghs)
```

```python
text = ' '.join(words)
# lower max_font_size
wordcloud = WordCloud(max_font_size=30, background_color='white').generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
